### Step 1: Importing Libraries

```python
import pandas
```

### Step 2: Importing dataset from wind.txt

```python
wd = pandas.read_csv('wind.txt', sep='\s+')
## for space separation reference
## https://stackoverflow.com/questions/19632075/how-to-read-file-with-space-separated-values-in-pandas
wd.head()
```

### Step 3: Assigning variable 'data' and replacing first 3 columns by datetime index.

```python
data = wd.copy()
date = pandas.to_datetime(data['Yr'] * 10000 + data['Mo'] * 100 + data['Dy'], format='%y%m%d')
data.insert(0, 'datetime', date)
data.drop(columns=['Yr', 'Mo', 'Dy'], inplace=True)
data
#https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html
```

## Step 4: Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
data['datetime'] = data['datetime'].apply( lambda x: pandas.Timestamp(year=x.year-100, month=x.month, day=x.day) if (x > pandas.Timestamp('2000-01-01')) else x)
data
##https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html --- timestamp
```

### Step 5: Setting the right dates as index with datetime64[ns] data type

```python
print(data['datetime'].dtype)
data.set_index('datetime', inplace=True)
data
```

### Step 6: Computing missing value for each location over the entire record.

```python
data.notna().sum().sum()
## https://pandas.pydata.org/docs/reference/api/pandas.notna.html
```

### Step 9: Creating DataFrame 'loc_stats' and calculating min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days

```python
loc_stats = data.apply(['min', 'max', 'mean', 'std'])
loc_stats
```

### Line plot for marriages and divorces per capita in the U.S. between 1867 and 2014

```python
import pandas
import matplotlib.pyplot as plt
```

```python
## Measurement Per Capita = Measurement / Population

df = pandas.read_csv('us-marriages-divorces-1867-2014.csv')
df['Marriages_per_capita'] = df['Marriages']/df['Population']
df['Divorces_per_capita'] = df['Divorces']/df['Population']
```

```python
plt.plot(df['Year'], df['Marriages_per_capita']) 
plt.plot(df['Year'], df['Divorces_per_capita'])
plt.xlabel('Years')
plt.ylabel('Cases Per Capita')
plt.legend(['Marriages per Capita', 'Divorces per Capita'])
```
